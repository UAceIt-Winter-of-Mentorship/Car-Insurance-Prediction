```python
from sklearn.impute import KNNImputer
from sklearn.impute import SimpleImputer

imputerm = SimpleImputer(strategy = 'most_frequent', missing_values = np.nan)
imputerm = imputerm.fit(dfc1[['communication']])
dfc1['communication'] = imputerm.transform(dfc1[['communication']])

imputerm = imputerm.fit(dfc1[['job_type']])
dfc1['job_type'] = imputerm.transform(dfc1[['job_type']])

imputerm = imputerm.fit(dfc1[['education_level']])
dfc1['education_level'] = imputerm.transform(dfc1[['education_level']])
dfc1.isnull().sum()
```

```python
imputerm = imputerm.fit(dfc[['job_type']])
dfc['job_type'] = imputerm.transform(dfc[['job_type']])

imputerm = imputerm.fit(dfc[['education_level']])
dfc['education_level'] = imputerm.transform(dfc[['education_level']])
dfc.isnull().sum()
```

```python
dfc['age'] = dfc['age'].astype(int)
dfc1['age'] = dfc1['age'].astype(int)
dfc.dtypes
```

```python
dfc['call_start'] = pd.to_datetime(dfc['call_start'])
dfc['call_end'] = pd.to_datetime(dfc['call_end'])
dfc['call_time'] = dfc['call_end'] - dfc['call_start']
dfc['call_time'] = dfc['call_time'].astype(np.int64)//10**9
dfc
```

```python
dfc1['call_start'] = pd.to_datetime(dfc1['call_start'])
dfc1['call_end'] = pd.to_datetime(dfc1['call_end'])
dfc1['call_time'] = dfc1['call_end'] - dfc['call_start']
dfc1['call_time'] = dfc1['call_time'].astype(np.int64)//10**9
dfc1
```

```python
train = dfc.drop(columns = ["communication", "call_start", "call_end", "last_contact_day", "last_contact_month"], inplace = False)
test = dfc1.drop(columns = ["communication", "call_start", "call_end", "last_contact_day", "last_contact_month"], inplace = False)
```

```python
print(train.dtypes, test.dtypes)
```

```python
train.head()
```

```python
job_type = pd.get_dummies(data=train['job_type'], drop_first=True)
marital_status = pd.get_dummies(data=train['marital_status'], drop_first=True)
education_level = pd.get_dummies(data=train['education_level'], drop_first=True)
train = pd.concat((job_type, marital_status, education_level,train), axis = 1 )
```

```python
del train['education_level']
del train['marital_status']
del train['job_type']
```

```python
pd.set_option("display.max_rows", 10, "display.max_columns", None)
```

```python
train.shape
```

```python
job_type = pd.get_dummies(data=test['job_type'], drop_first=True)
marital_status = pd.get_dummies(data=test['marital_status'], drop_first=True)
education_level = pd.get_dummies(data=test['education_level'], drop_first=True)
test = pd.concat((job_type, marital_status, education_level,test), axis = 1 )
```

```python
pd.set_option("display.max_rows", None, "display.max_columns", None)
test.dtypes, test.shape
```

```python
test.drop(columns = ["job_type", "marital_status", "education_level"], inplace = True)
```

```python
list(set(train.columns) - set(test.columns))[0]
```

```python
y = train["car_insurance"].values
X = train.drop(columns = "car_insurance", inplace = False)
X = X.values
```

```python
test.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state = 100)
print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)
```

```python
from sklearn import tree
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
clfdt = tree.DecisionTreeClassifier()
clfdt.fit(X_train, y_train)
preddt = clfdt.predict(X_test)
print(clfdt.score(X_test, y_test))
print(classification_report(y_test, preddt))
print(confusion_matrix(y_test, preddt))
```

```python
from sklearn.neighbors import KNeighborsClassifier
clfkn = KNeighborsClassifier(n_neighbors = 10)
clfkn.fit(X_train, y_train)
predkn = clfkn.predict(X_test)
print(clfkn.score(X_test, y_test))
print(classification_report(y_test, predkn))
print(confusion_matrix(y_test, predkn))
```

```python
from sklearn.linear_model import LogisticRegression
clflr = LogisticRegression(random_state = 0)
clflr.fit(X_train, y_train)
predlr = clflr.predict(X_test)
print(clflr.score(X_test, y_test))
print(classification_report(y_test, predlr))
print(confusion_matrix(y_test, predlr))
```

```python
from sklearn.ensemble import RandomForestClassifier
clfrf = RandomForestClassifier(random_state = 0, n_estimators=110)
clfrf.fit(X_train, y_train)
predrf = clfrf.predict(X_test)
print(clfrf.score(X_test, y_test))
print(classification_report(y_test, predrf))
print(confusion_matrix(y_test, predrf))
```

```python
predfrf= clfrf.predict(test)
prediction = pd.DataFrame(predfrf, columns=["prediction"])
prediction.to_csv("Prediction.csv", index = False)
```
